```python
X = data[[ 'Q2-出生年', 'Q3-性别', 'Q4-民族', 'Q6-文化程度', 'Q7-婚姻状况', 'Q8-职业']]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs


# 计算不同聚类个数对应的平均误差（Inertia）
inertias = []
for k in range(1, 20):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# 绘制肘部法图形
plt.plot(range(1, 20), inertias, marker='o')
plt.xlabel('聚类个数 (k)')
plt.ylabel('迭代')
plt.title('肘部迭代图')
plt.savefig('./Q4/肘部迭代图')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
# 使用 PCA 进行降维到二维空间
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 创建 K-Means 聚类模型
kmeans = KMeans(n_clusters=3, random_state=42)

# 使用降维后的数据拟合模型
kmeans.fit(X_pca)

# 预测数据所属的簇
labels = kmeans.predict(X_pca)

# 获取聚类中心点的坐标
centroids = kmeans.cluster_centers_

# 可视化聚类结果
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
plt.legend()
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering with PCA')
plt.savefig('./Q4/K-Means')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
 
plt.rcParams['font.sans-serif'] = ['SimHei']  
# Matplotlib中设置字体-黑体，解决Matplotlib中文乱码问题
plt.rcParams['axes.unicode_minus'] = False    
# 解决Matplotlib坐标轴负号'-'显示为方块的问题
sns.set(font='SimHei')                        
# Seaborn中设置字体-黑体，解决Seaborn中文乱码问题

# 使用Seaborn绘制频率分布直方图
#https://jingyan.baidu.com/article/eb9f7b6d61838cc79264e86a.html
import pandas as pd

def sigma(series_data):  
    # 计算数据的均值和标准差
    mean_val = series_data.mean()
    std_dev = series_data.std()

    # 定义异常值的阈值范围
    threshold = 3 * std_dev

    # 进行3σ异常值清洗
    return series_data[series_data.between(mean_val - threshold, mean_val + threshold)]
    
    
for i in data[['谷薯类每天食用次数', '谷薯类每天食用次数（周/7）', '畜禽鱼蛋奶', '畜禽鱼蛋奶（周/7）',
       '豆类食物', '豆类食物（周/7）', '蔬菜水果', '蔬菜水果（周/7）', '每天摄入食物类型数量', '每周摄入食物类型数量',
       'BMI', 'BMI_Category', '每周参加体育锻炼频率', '平均每天体育锻炼时间', '每天食用次数总额',
       '菜水果、全谷物和奶制品占比每天摄入频率比例', '新鲜蔬菜摄入量', '水果摄入量', '各种各样的奶制品摄入量',
       '全谷物和奶制品每周摄入次数', '鱼禽、蛋类和瘦肉每天摄入量', '每周吃水产品频率', '每天吃蛋类个数', '每天食用肉类频率总额',
       '食用水产的频率占比每天食用肉类频率比例']]:
    aaa=sigma(data[i])
    sns.distplot(aaa, color='Pink')
    display(aaa.describe())
    # 添加标题和标签
    # plt.title('每周摄入食物类型数量')
    plt.xlabel(i)
    plt.ylabel('Frequency')
    # 显示图形
    plt.savefig('./Q4/%s'%i.replace('/',''))
    plt.show()
    
```
